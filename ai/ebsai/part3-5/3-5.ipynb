```python
def AND(x1, x2):
    w1, w2, threshold = 0.2, 0.2, 0.3
    temp = w1 * x1 + w2 * x2
    if temp <= threshold:
        return 0
    else:
        return 1

print(AND(2, 0))
print(AND(1, 0))
```

```python
def OR(x1, x2):
    w1, w2, threshold = 0.3, 0.3, 0.2
    temp = w1 * x1 + w2 * x2
    if temp <= threshold:
        return 0
    else:
        return 1
```
