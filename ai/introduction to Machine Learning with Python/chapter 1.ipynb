```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
```

```python
print(X_train.shape)
```

```python
print(y_train.shape)
```

```python
import pandas as pd
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
```

```python
import mglearn
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='x', hist_kwds={'bins':20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
```

```python
knn.fit(X_train, y_train)
```

```python
knn
```

```python
import numpy as np
X_new = np.array([[5, 2.9, 1, 0.2]])
```

```python
print('X_new.shape', X_new.shape)
```

```python
prediction = knn.predict(X_new)
```

```python
print('예측:', prediction)
print('에측한 타깃의 이름:', iris_dataset['target_names'][prediction])
```

```python
y_pred = knn.predict(X_test)
print('테스트 세트에 대한 예측값:\n', y_pred)
```

```python
print("테스트 세트의 정확도: {:.3f}".format(np.mean(y_pred == y_test)))
```
